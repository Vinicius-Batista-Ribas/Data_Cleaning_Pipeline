# Data Cleaning Pipeline
# ============================================

Pipeline de limpeza e padronização de dados aplicado ao dataset do catálogo da Netflix.

O processo inclui tratamento de valores nulos, remoção de duplicatas, padronização de categorias, transformação de tipos de dados e criação de novas variáveis para análise exploratória.

# 1) Importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2) Carregar dataset

```python
df = pd.read_csv("NetFlix.csv")
```

# 3) Exploração inicial do dataset

```python
# verificar primeiras linhas
df.head()
```

```python
# verificar estrutura dos dados
df.info()
```

```python
# verificar estatísticas básicas
df.describe()
```

```python
# verificar valores nulos
df.isnull().sum()
```

# 4) Remover duplicatas

```python
# verificar se existem
df.duplicated().sum()
```

# 5) Tratar valores nulos

```python
# diretor ausente
df["director"] = df["director"].fillna("Unknown")

# elenco ausente
df["cast"] = df["cast"].fillna("Not informed")

# país ausente
df["country"] = df["country"].fillna("Unknown")

# rating ausente
df["rating"] = df["rating"].fillna("Not Rated")

# remover linhas com data ausente (poucas)
df = df.dropna(subset=["date_added"])
```

# 6) Corrigir tipos de dados

```python
# converter coluna de data
df["date_added"] = pd.to_datetime(df["date_added"])
```

# 7) Padronizar valores de texto

```python
# remover espaços extras
df["country"] = df["country"].str.strip()

# padronizar formato
df["country"] = df["country"].str.title()

# padronizar gêneros
df["genres"] = df["genres"].str.strip()
```

# 8) Separar colunas compostas

```python
# pegar apenas o primeiro país listado
df["country_main"] = df["country"].str.split(",").str[0]

# criar colunas para cada gênero
genres_dummies = df["genres"].str.get_dummies(sep=", ")

# juntar com o dataset original
df = pd.concat([df, genres_dummies], axis=1)

# deletar coluna de generos antiga
df = df.drop(columns=["genres"])
```

# 9) Criar novas variáveis (feature engineering)

```python
# ano que foi adicionado na plataforma
df["year_added"] = df["date_added"].dt.year

# mês que foi adicionado
df["month_added"] = df["date_added"].dt.month
```

# 10) Validar dados após limpeza

```python
# verificar novamente nulos
df.isnull().sum()
```
